Write a program to read through a file and output the number of lines in a file.

adapted from [PEW]

```python
f = open('mary.txt')
lines = f.readlines()
print(len(lines))
```

```python
try:
    f = 'mary.txt'
    with open(f) as file:
      lines = file.readlines()
      print('{} has {} lines.'.format(f, len(lines)))
except:
      print(f+ ' does not exist.')
```
