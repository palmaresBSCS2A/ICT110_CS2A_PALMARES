```python
%%writefile capiz.py

"""
This module provides functions related to Capiz town names and their postal codes.
"""

capiz_dict = {
    'Ivisan': 5805,
    'Roxas City': 5800,
    'Dumarao': 5812,
    'Sapian': 5806,
    'Tapaz': 5814
}

def postal_code(place_name):
    """
    Retrieve the postal code for a given place name from the capiz_dict dictionary.

    Parameters: the name of the place for which to retrieve the postal code.

    Returns: the postal code of the place if found in the dictionary, otherwise returns None.
    """
    return capiz_dict.get(place_name)
```

```python
import capiz
```

```python
place = input('Enter Town in Capiz: ')
code = capiz.postal_code(place)
print(f"Postal Code of {place} is {code}.")
```
