B. Get the maximum number for each dimensions and display it. Explain what is the result briefly by writing a comment before the code.

```python
# the result is 12 since the random generated highest number in the 1st dimension is 12
d1 = np.max(x4, axis=0).max()
d1
```

```python
# the result is 12 since the random generated highest number in the 2nd dimension is 12
d2 = np.max(x4, axis=1).max()
d2
```

```python
# the result is 12 since the random generated highest in number the 3rd dimension is 12
d3 = np.max(x4, axis=2).max()
d3
```

```python
# the result is 12 since the random generated highest number in the 4th dimension is 12
d4 = np.max(x4, axis=3).max()
d4
```
