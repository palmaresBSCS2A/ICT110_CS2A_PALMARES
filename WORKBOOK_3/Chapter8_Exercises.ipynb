## CHALLENGE

Modify the code in the subheading [Parsing Lines](#ParsingLines) to display all the time stamps
of the emails in mboxt-short.txt


```python
fhand = open('/content/drive/MyDrive/WORKBOOK_3/mbox-short.txt')
timestamps = []
for line in fhand:
    line = line.rstrip()
    words = line.split()
    for word in words:
        if len(word) == 8:
              time = word.split(':')
              if len(time) == 3:
                hours = int(time[0])
                minutes = int(time[1])
                seconds = int(time[2])
                if 0<= hours <24 and 0 <=minutes <60 and 0<=seconds<60:
                  timestamps.append(word)
for timestamp in timestamps:
    print(timestamp)
```
